```python
from numpy import ones,vstack
from numpy.linalg import lstsq
import pandas as pd
import numpy as np
np.warnings.filterwarnings('ignore')
```

```python
year2020 = pd.read_csv("ratings/withRatings_stats_2020.csv")
year2021 = pd.read_csv("ratings/withRatings_stats_2021.csv")
year2022 = pd.read_csv("ratings/withRatings_stats_2022.csv")
```

```python
rating_2021 = []
totalGames = []
for index,row in year2022.iterrows():
    try:
        pastRating = year2021.loc[year2021['Player'] == (row.Player)].scaledplayerRating.item()
        pastGP = year2021.loc[year2021['Player'] == (row.Player)].G.item()
        rating_2021.append(pastRating)
        totalGames.append(row.G + pastGP)

    except:
        print("No data found for " + row.Player)
        rating_2021.append(row.scaledplayerRating)
        totalGames.append(row.G)


year2022["rating_2021"] = rating_2021
year2022["threeYearGames"] = totalGames

```

```python
rating_2020 = []
totalGames = []
for index,row in year2022.iterrows():
    try:
        pastRating = year2020.loc[year2020['Player'] == (row.Player)].scaledplayerRating.item()
        pastGP = year2020.loc[year2020['Player'] == (row.Player)].G.item()
        rating_2020.append(pastRating)
        totalGames.append(row.threeYearGames + pastGP)
    except:
        print("No data found for " + row.Player)
        rating_2020.append(row.scaledplayerRating)
        totalGames.append(row.threeYearGames)

year2022["rating_2020"] = rating_2020
year2022["threeYearGames"] = totalGames
```

```python
missedGamesWeight = {
    "weight2020": .25,
    "weight2021": 1.975,
    "weight2022": 0.775
}

regularGamesWeight = {
    "weight2020": .15,
    "weight2021": .475,
    "weight2022": 2.375
} 


avgRating = []
for index, row in year2022.iterrows():
    if row.G < 40:
        avgR = int(((row.rating_2020 * missedGamesWeight["weight2020"]) + (row.rating_2021 * missedGamesWeight["weight2021"]) + (row.scaledplayerRating * missedGamesWeight["weight2022"])) / 3)
    else:
        avgR = int(((row.rating_2020 * regularGamesWeight["weight2020"]) + (row.rating_2021 * regularGamesWeight["weight2021"]) + (row.scaledplayerRating * regularGamesWeight["weight2022"])) / 3)
    avgRating.append(avgR)

year2022["avgRating"] = avgRating
```

```python
sortedBy = year2022.sort_values(by=["avgRating"], ascending=False)

for index, row in sortedBy.iterrows():
    string = "{:>24} | Rating: {} | Off Rating: {} | Def Rating: {} | Avg Rating: {} | Games Played: {} | Rating Difference: {}"
    print(string.format(row.Player,str(row.scaledplayerRating),str(row.offensiveRating),str(row.defensiveRating),str(row.avgRating),str(row.G), str(row.avgRating - row.scaledplayerRating)))
```
