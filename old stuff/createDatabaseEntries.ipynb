```python
import config
from numpy import ones,vstack
from numpy.linalg import lstsq
import pandas as pd
import numpy as np
np.warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('finalRatings/fr_2022.csv')
```

```python
import unicodedata

def strip_accents(text):
    return ''.join(c for c in unicodedata.normalize('NFKD', text) if unicodedata.category(c) != 'Mn')
```

```python
for index, row in df.iterrows():
    df.loc[index, "PlayerName"] = strip_accents(row.PlayerName)
```

```python
df2 = pd.read_csv("NBA_PLAYERS.csv")
```

```python
HT = []
LAT = []
LNG = []
finalData = []
for index,row in df.iterrows():
    try:
        player = df2.loc[(df2['FIRST_NAME'] + " " + df2['LAST_NAME']) == (row.PlayerName)]
        finalData.append([
            player.PLAYER_ID.item(), 
            player.FIRST_NAME.item(), 
            player.LAST_NAME.item(), 
            player.AGE.item(),
            player.HOMETOWN.item(),
            player.LONGITUDE.item(),
            player.LATITUDE.item(),
            row.scaledRatingSum,
            row.Position
            ]
            )
    except:
        print("No data found for " + row.PlayerName)

newDF = pd.DataFrame(finalData, columns=["PLAYER_ID",
                                         "FIRST_NAME",
                                         "LAST_NAME",
                                         "AGE",
                                         "HOMETOWN",
                                         "LONGITUDE",
                                         "LATITUDE",
                                         "RATING",
                                         "POS"
                                         ])

```

```python
newDF.to_csv("finalDatabaseExcel.csv")
```
