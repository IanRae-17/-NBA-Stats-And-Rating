```python
import requests
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
years = list(range(2022, 2023))
conference = ["E", "W"]
for year in years:
    print(year)
```

```python
url_start = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
for year in years:
    url = url_start.format(year)
    data = requests.get(url)

    with open("standings/{}_standings.html".format(year), "w+", encoding="utf-8") as f:
        f.write(data.text)
```

```python
standings = {}

for c in conference:
    idName = "confs_standings_" + c
    for year in years:
        with open("standings/{}_standings.html".format(year), encoding="utf-8") as f:
            page = f.read()

            soup = BeautifulSoup(page, "html.parser")
            totals = soup.find(id=idName)
            standings[str(year) + "_" + c] = pd.read_html(str(totals))[0]
```

```python

data = []

for x in standings:
    standingsHeader = standings[x].columns[0]
    for index, row in standings[x].iterrows():
        data.append([getAbbr(row[standingsHeader].replace("*", "")), row["W/L%"]])

standings = pd.DataFrame(data, columns=['team', 'winPC'])

display(standings)
```
